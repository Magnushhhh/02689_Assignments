```python
import numpy as np
import matplotlib.pyplot as plt
from JacobiP import *
#from scipy.special import *
```

```python
def JacobiP(x,alpha,beta,n):
    if n == 0:
        return np.ones(len(x))
    elif n == 1:
        return 1/2*(alpha-beta+(alpha+beta+2)*x)
    else:
        a1 = (2*(n+alpha)*(n+beta))/((2*n+alpha+beta+1)*(2*n+alpha+beta))
        ann = (alpha**2-beta**2)/((2*n+alpha+beta+2)*(2*n+alpha+beta))
        a2 = (2*(n+1)*(n+alpha+beta+1))/((2*n+alpha+beta+2)*(2*n+alpha+beta+1))
        return ((ann+x)*JacobiP(x,alpha,beta,n-1)-a1*JacobiP(x,alpha,beta,n-2))/a2

# Legendre    
N = 100
xs = np.array([i*2/N-1 for i in range(N+1)])
plt.figure()
for n in range(6):
    P = JacobiP(x=xs,alpha=0,beta=0,n=n)
    plt.plot(xs,P)

plt.show()

# Chebyshe # gamma doesn't work right now
# plt.figure()
# for n in range(6):
#     P = JacobiP(x=xs,alpha=-1/2,beta=-1/2,n=n)
#     T = scp.special.gamma(n+1)*scp.special.gamma(1/2)/(scp.special.gamma(n+1/2))*P
#     plt.plot(xs,T)

# plt.show()

```
